```python
%matplotlib inline
import torch
from torch import nn
from d2l import torch as d2l
```

### 数据生成

$y = 0.05 + \sum_{i=1}^d 0.01x_i + \epsilon~where~\epsilon ~ \sim N(0,0.01^2)$

```python
n_train, n_test, num_inputs, batch_size = 20, 100, 200, 5 # 训练样本小，容易过拟合，输入维度是 200
true_w, true_b = torch.ones((num_inputs, 1)) * 0.01, 0.05 # true_w 是 num_inputs * 1
train_data = d2l.synthetic_data(true_w, true_b, n_train)
train_iter = d2l.load_array(train_data, batch_size)
test_data = d2l.synthetic_data(true_w, true_b, n_test)
test_iter = d2l.load_array(test_data, batch_size, is_train=False)
```

### 初始化模型参数

```python
def init_params():
    w = torch.normal(0, 1, size=(num_inputs, 1), requires_grad=True)
    b = torch.zeros(1, requires_grad=True)
    return [w, b]
```

### 定义惩罚范数

```python
def l2_penalty(w):
    return torch.sum(w.pow(2)) / 2
```

### 训练函数

```python
def train(lambd):
    w, b = init_params()
    net, loss = lambda X: d2l.linreg(X, w, b), d2l.squared_loss # 线性回归模型，均方误差
    num_epochs, lr = 100, 0.003
    animator = d2l.Animator(xlabel='epoches', ylabel='loss', yscale='log', xlim=[5, num_epochs], legend=['train', 'test'])
    for epoch in range(num_epochs):
        for X, y in train_iter:
            l = loss(net(X), y) + lambd * l2_penalty(w) # lambd 是超参数
            l.sum().backward()
            d2l.sgd([w, b], lr, batch_size) # 自己写的 SGD
        if (epoch + 1) % 5 == 0:
            animator.add(epoch + 1, (d2l.evaluate_loss(net, train_iter, loss), d2l.evaluate_loss(net, test_iter, loss)))
    print('w 的 L2 范数是：', torch.norm(w).item())
```

### 训练

忽略正则化直接训练，出现了严重的过拟合

```python
train(lambd=0)
```

使用正则化（`lambd = 3`）训练，有一定过拟合，但好了很多，范数较小（w 较小）

```python
train(lambd=3)
```

```python
train(lambd=10)
```

### 简洁实现

```python
def train_consice(wd):
    loss = nn.MSELoss(reduction='none') # 均方损失
    net = nn.Sequential(nn.Linear(num_inputs, 1)) # 网络
    for params in net.parameters():
        params.data.normal_() # 初始化 w
    num_epoches, lr = 100, 0.003
    trainer = torch.optim.SGD([{"params":net[0].weight, "weight_decay":wd}, {"params":net[0].bias}], lr=lr) # 罚函数在自定义 SGD 参数中实现
    animator = d2l.Animator(xlabel='epoch', ylabel='loss', yscale='log', xlim=[5, num_epoches], legend=['train', 'test'])
    for epoch in range(num_epoches):
        for X, y in train_iter:
            trainer.zero_grad()
            l = loss(net(X), y)
            l.mean().backward()
            trainer.step()
        if (epoch + 1) % 5 == 0:
            animator.add(epoch + 1, (d2l.evaluate_loss(net, train_iter, loss), d2l.evaluate_loss(net, test_iter, loss)))
    print('w 的 L2 范数是：', net[0].weight.norm().item())
```

```python
train_consice(wd=0)
```

```python
train_consice(wd=10)
```

### 总结

1. 简介实现中，在 SGD 中自定义参数，在需要衰退的参数后加 `weight_decay`:weight_decay（lambda）
2. d2l 提供的画图函数，`xlabel, ylabel` 是 x，y 轴的名称，`xscale, yscale` 是以指数显示范围，目前知道 `'log'`，`xlim[start, end], ylim[start, end]` 是指定范围
